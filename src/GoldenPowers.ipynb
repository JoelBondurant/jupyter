# Golden Powers
---

```python
import numpy as np
from mpmath import mp
import matplotlib as mpl, matplotlib.pyplot as plt
%matplotlib inline
```

```python
mp.prec = 6000
N0 = 0
N = 1000
x = np.arange(N0, N)
y = [mp.power(mp.phi, n) for n in range(N0, N)]
```

```python
y2 = [mp.fsub(yp, mp.nint(yp)) for yp in y]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8,8))
ax.plot(x, y2, color = '#e1e114')
ax.set_title('Golden Powers')
ax.set_xscale('log')
#ax.set_yscale('log')
yl = 1e-9
ax.set_ylim(-yl, yl)
ax.set_xlabel('$n$')
ax.set_ylabel('$φ^n - nint(φ^n)$\n$φ = Golden Ratio$')
ax.grid()
plt.tight_layout()
#plt.savefig('golden_powers.jpg', dpi = 300);
```
