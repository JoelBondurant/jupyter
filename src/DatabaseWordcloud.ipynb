```python
import wikipedia
import wordcloud
import matplotlib.pyplot as plt
```

```python
wc = wordcloud.WordCloud(width = 2000, height = 1000, color_func=wordcloud.get_single_color_func('#0844aa')).generate(wikipedia.page('database').content)
```

```python
fig = plt.figure(figsize = (20,10))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
