# Jacobi j-Theta Function
---

```python
import numpy as np
import scipy.special
import sympy as sp
import matplotlib as mpl, matplotlib.pyplot as plt
%matplotlib inline
```

```python
sp.mpmath.fp.cplot(lambda z: sp.mpmath.jtheta(1, z, np.exp(1j*np.pi/10)/10), [-10,10], [-10,10], points = 1e5, dpi = 128)
```

```python
def mkdata(ll = 10):
    re = (-ll,ll)
    im = (-ll,ll)
    n = 92*8
    x = np.linspace(re[0], re[1], n)
    y = np.linspace(im[0], im[1], n)
    xx, yy = np.meshgrid(x, y)
    z = (np.vectorize(lambda zv: complex(sp.mpmath.jtheta(1, zv, np.exp(1j*np.pi/10)/10)))(xx + 1j*yy))
    zgray = 2*np.arcsin(np.absolute(np.imag(z)/z))/np.pi
    print((len(x),len(y),len(z),len(z[0]),xx.shape,yy.shape, zgray.shape))
    return x, y, zgray

def mkdata2(nn, mm, ll = 10):
    re = (-ll,ll)
    im = (-ll,ll)
    n = 92*8
    x = np.linspace(re[0], re[1], n)
    y = np.linspace(im[0], im[1], n)
    xx, yy = np.meshgrid(x, y)
    z = (np.vectorize(lambda zv: complex(sum([sp.mpmath.jtheta(nnn, zv, np.exp(1j*np.pi/10)/10) for nnn in range(nn, mm+1)])))(xx + 1j*yy))
    zgray = 2*np.arcsin(np.absolute(np.imag(z)/z))/np.pi
    print((len(x),len(y),len(z),len(z[0]),xx.shape,yy.shape, zgray.shape))
    return x, y, zgray
```

```python
def mkplot(xyz, nn = 10, cmap = mpl.cm.Blues):
    plt.rcParams['figure.figsize'] = (8,8)
    plt.imshow(xyz[2], extent = (np.amin(xyz[0]), np.amax(xyz[0]), np.amin(xyz[1]), np.amax(xyz[1])), cmap = cmap, alpha = 1)
    plt.tight_layout()
    plt.savefig('jtheta%s.jpg' % nn, dpi = 300)
```

```python
xyz = mkdata()
mkplot(xyz)
```

```python
mkplot(xyz, cmap = mpl.cm.BuPu_r)
```

```python
mkplot(xyz, cmap = mpl.cm.jet)
```

```python
xyz = mkdata2(1,4)
mkplot(xyz, cmap = mpl.cm.BuPu_r)
```
