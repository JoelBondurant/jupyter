```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('ai.csv', skiprows = 1)
df.columns = ('Month','Artificial_Intelligence')
df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (16,6))
ax.plot(df.index, df.Artificial_Intelligence, label = 'Google Search Volume', lw = 3, alpha = 1)
ticks = range(0, len(df.Month), 6)
tick_labels = df.Month.iloc[ticks]
ax.xaxis.set_ticks(ticks)
ax.xaxis.set_ticklabels(tick_labels)
[tl.set_rotation(24) for tl in ax.xaxis.get_ticklabels()]
ax.set_title('Artificial Intelligence', fontsize = 16)
ax.legend()
ax.grid()
plt.tight_layout()
plt.savefig('ai.png', dpi = 300)
```
