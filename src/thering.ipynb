```python
import numpy as np
import matplotlib as mpl, matplotlib.pyplot as plt
%matplotlib inline
```

```python
def kb(z, n):
    return z * np.exp(2*np.pi*1j*n/3)

def ka(z, n):
    return z/n

def kfrac(z):
    return kb(z, 1)/(ka(z,1)+kb(z,2)/(ka(z,2)+kb(z,3)/(ka(z,3)+kb(z,4)/(ka(z,4)+kb(z,5)/(ka(z,5)+kb(z,6)/(ka(z,6)+kb(z,7)/(ka(z,7)+kb(z,8)/(ka(z,8)+kb(z,9)/(ka(z,9)+kb(z,10)/ka(z,10))))))))))
```

```python
re = (-6,3)
im = (-3,6)
n = 55**2
x = np.linspace(re[0], re[1], n)
y = np.linspace(im[0], im[1], n)
xx, yy = np.meshgrid(x, y)
z = kfrac(xx + 1j*yy)
zgray = 2*np.arcsin(np.absolute(np.imag(z)/z))/np.pi
print((len(x),len(y),len(z),len(z[0]),xx.shape,yy.shape, zgray.shape))
```

```python
plt.rcParams['figure.figsize'] = (8,8)
plt.imshow(zgray, extent = (np.amin(x), np.amax(x), np.amin(y), np.amax(y)), cmap = mpl.cm.gray_r, alpha = 1)
plt.tight_layout()
```
