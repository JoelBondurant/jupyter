```rust
:dep reqwest = { features = ["blocking"] }
```

```rust
:dep chrono
use chrono::prelude::*;
```

```rust
:dep plotters = { default_features = false, features = ["evcxr", "all_series", "datetime"] }
use plotters::prelude::*;
```

```rust
let url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=RESPPLLOPNWW";
let csv_txt = reqwest::blocking::get(url)?.text()?;
```

```rust
let csv_data = csv_txt.split("\n").map(|x| {
    x.split(",").map(|x| x.to_owned()).collect::<Vec<_>>()
}).filter(|x| x.len() == 2).collect::<Vec<_>>();
```

```rust
let data = csv_data[1..].iter().map(|x| 
    (x[0].parse::<NaiveDate>().unwrap(), x[1].parse::<f32>().unwrap().floor() as i32)
).collect::<Vec<_>>();
```

```rust
evcxr_figure((840, 420), |root| {
    root.fill(&WHITE)?;
    let start_date = "2010-12-01".parse::<NaiveDate>().unwrap();
    let end_date = "2023-01-01".parse::<NaiveDate>().unwrap();
    let mut chart = ChartBuilder::on(&root)
        .caption("FED Liabilities: Earnings Remittances Due to U.S. Treasury", ("sans-serif", 20))
        .set_label_area_size(LabelAreaPosition::Left, 60)
        .set_label_area_size(LabelAreaPosition::Bottom, 30)
        .build_cartesian_2d(start_date..end_date, -12000..6000)?;
    chart.configure_mesh().draw()?;
    chart.draw_series(LineSeries::new(data.clone(), ShapeStyle {
        color: RGBAColor(0, 80, 200, 8.0), filled: false, stroke_width: 2 }))?;
    Ok(())
})
```
