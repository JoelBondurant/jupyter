```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('merica_time.csv')
df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (16,6))
for i in range(1, 7):
    ax.plot(df.Age, df.iloc[:, i], label = df.columns[i], lw = 3, alpha = 1)
ax.yaxis.set_ticks(range(0, 10, 1))
ax.xaxis.set_ticks(range(15, 86, 5))
ax.set_ylim((0, 9))
ax.set_xlim((15, 85))
ax.set_ylabel('hours/day', fontsize = 14)
ax.set_xlabel('age', fontsize = 14)
ax.set_title('', fontsize = 16)
ax.legend()
ax.grid()
```
