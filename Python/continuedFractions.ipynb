```python
import numpy as np
import matplotlib as mpl, matplotlib.pyplot as plt
%matplotlib inline
```

```python
def kb(z, n):
    return np.sinh(z)*z* np.exp(2*np.pi*1j*n) 

def ka(z, n):
    return np.cos(z)/n
```

```python
def kfrac(z):
    return kb(z, 1)/(ka(z,1)+kb(z,2)/(ka(z,2)+kb(z,3)/(ka(z,3)+kb(z,4)/(ka(z,4)+kb(z,5)/(ka(z,5)+kb(z,6)/(ka(z,6)+kb(z,7)/(ka(z,7)+kb(z,8)/(ka(z,8)+kb(z,9)/(ka(z,9)+kb(z,10)/ka(z,10))))))))))
```

```python
def mkdata():
    re = (-10,10)
    im = (-10,10)
    n = 55**2
    x = np.linspace(re[0], re[1], n)
    y = np.linspace(im[0], im[1], n)
    xx, yy = np.meshgrid(x, y)
    z = kfrac(xx + 1j*yy)
    zgray = 2*np.arcsin(np.absolute(np.imag(z)/z))/np.pi
    print((len(x),len(y),len(z),len(z[0]),xx.shape,yy.shape, zgray.shape))
    return x, y, zgray
```

```python
def mkplot(xyz, nn = 8, cmap = mpl.cm.Blues):
    plt.rcParams['figure.figsize'] = (8,8)
    plt.imshow(xyz[2], extent = (np.amin(xyz[0]), np.amax(xyz[0]), np.amin(xyz[1]), np.amax(xyz[1])), cmap = cmap, alpha = 1)
    plt.tight_layout()
    plt.savefig('contfrac_%s.jpg' % nn, dpi = 300)
```

```python
xyz = mkdata()
mkplot(xyz)
```

```python
mkplot(xyz, cmap = mpl.cm.hot)
```

```python
mkplot(xyz, cmap = mpl.cm.hot_r)
```

```python
def kb6(z, n):
    return z**2 * np.exp(2*np.pi*1j*n)

def ka6(z, n):
    return np.cosh(z)/n
```

```python
def kb7(z, n):
    return np.sin(z)*z * np.exp(2*np.pi*1j*n) 

def ka7(z, n):
    return np.cos(z)/n
```
