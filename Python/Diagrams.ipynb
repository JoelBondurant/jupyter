```python
import urllib
from PIL import Image
from diagrams import Diagram, Cluster
from diagrams.custom import Custom
from diagrams.onprem.workflow import Airflow
from diagrams.gcp.analytics import BigQuery, Composer
from diagrams.gcp.storage import GCS
```

```python
custom_urls = [
    ("airflow.png", "https://static-00.iconduck.com/assets.00/airflow-icon-2048x2048-ptyvisqh.png"),
    ("talend.png", "https://upload.wikimedia.org/wikipedia/commons/7/70/TalendLogoCoral.png"),
]
for fname, furl in custom_urls:
    with urllib.request.urlopen(furl) as fin, open(fname, "wb") as fout:
        fout.write(fin.read())
Image.open("talend.png").crop((80,110,464,494)).save("talend.png")
```

```python
with Diagram("diagram_name", show=False, ) as diag:
    with Cluster("cluster_name") as clus:
        a = GCS("gcs_name")
        b = BigQuery("bq_name")
        c1 = Composer("c_name\nline2") 
        c2a = Custom("a1_name", "talend.png")
        c2b = Airflow("a2_name")
        a >> c1 >> c2a >> b
        c1 >> c2b >> b
diag
```
