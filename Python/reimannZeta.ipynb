```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
import scipy.special
```

```python
fig = plt.figure(figsize = [14, 5])
ax = fig.add_subplot(111)

ax.set_title('Riemann Zeta Function', fontsize = 18)
ax.grid(which = 'major')
ax.minorticks_on()

zeta = lambda x: scipy.special.zeta(x, 1)

z = pd.DataFrame({'x': np.linspace(1.1, 4, 10**5)})
z['y'] = z.x.map(zeta)

ax.plot(z.x, z.y, linestyle = '-', linewidth = 4, color = '#0083bf')
```
