## UdpSocket1
---

```rust
:dep tokio = { features = ["full"] }
```

```rust
async fn udp_listener() {
    let net_mask = ":::7901";
    println!("Listening on {net_mask}.");
    let sock = tokio::net::UdpSocket::bind(net_mask).await.unwrap();
    let mut buf = [0; 1024];
    loop {
        let (len, addr) = sock.recv_from(&mut buf).await.unwrap();
        let msg = std::str::from_utf8(&buf).unwrap();
        println!("{len} bytes received from {addr}:\n{msg}");
    }
}
```

```rust
udp_listener().await
```
